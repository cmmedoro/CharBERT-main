## NER

```python
URLS = {
    "charbertwiki": "https://drive.google.com/file/d/1vQK8_20PiRfhmhqH4dfW0XeE11_pZ2FT/view?usp=sharing",

}

import os
import gdown
import shutil


os.makedirs("data/model_pretrained", exist_ok=True)
for dataset_name, url in URLS.items():
    print(f"Downloading {dataset_name}")
    zip_filepath = f"data/{dataset_name}.zip"
    gdown.download(url, zip_filepath, fuzzy=True)
    shutil.unpack_archive(zip_filepath, extract_dir="data")
    os.remove(zip_filepath)
```
